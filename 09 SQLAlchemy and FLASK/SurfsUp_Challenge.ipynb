```python
# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect



#import additional dependencies
import datetime as dt
import pandas as pd
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Get the name of the table. 
inspector=inspect(engine)
inspector.get_table_names()

#Get columns in measurement table
columns=inspector.get_columns('measurement')
for column in columns:
    print(column["name"], column["type"])
    
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
#took out 2017 year filter....need to check if this is accurate as it would give you all June observations regardless of year.
session.query(Measurement.tobs)
june=session.query(Measurement.tobs).\
filter(extract('year', Measurement.date)).\
filter(extract('month', Measurement.date)==6).all()
print(june)
```

```python
#  3. Convert the June temperatures to a list.
june_temps = list(np.ravel(june))
june_temps    
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_temps, columns=['June Temps'])
print(june_df)
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
#took out 2016 year filter....need to check if this is accurate as it would give you all June observations regardless of year.
session.query(Measurement.tobs)
december=session.query(Measurement.tobs).\
filter(extract('year', Measurement.date)).\
filter(extract('month', Measurement.date)==12).all()
print(december)
```

```python
# 7. Convert the December temperatures to a list.
december_temps = list(np.ravel(december))
december_temps    
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_df = pd.DataFrame(december_temps, columns=['Dec Temps'])
print(december_df)
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
december_df.describe()
```

# **D3: Additional Queries**

```python
#percipiation data for June
session.query(Measurement.prcp)
june=session.query(Measurement.prcp).\
filter(extract('year', Measurement.date)).\
filter(extract('month', Measurement.date)==6).all()
#June percipiation to a list.
june_prcp = list(np.ravel(june))
#Create a DataFrame from the list of percipiation for the month of June. 
june_prcp_df = pd.DataFrame(june_prcp, columns=['June Percipiation'])
#Detailed stats
june_prcp_df.describe()
```

```python
#percipiation data for December
session.query(Measurement.prcp)
december=session.query(Measurement.prcp).\
filter(extract('year', Measurement.date)).\
filter(extract('month', Measurement.date)==12).all()
#December percipiation to a list.
december_prcp = list(np.ravel(december))
#Create a DataFrame from the list of percipiation for the month of December. 
december_prcp_df = pd.DataFrame(december_prcp, columns=['December Percipitation'])
#Detailed stats
december_prcp_df.describe()
```

```python
#temperature & percipitation stats

temps_june_stat=june_df.describe()
temps_dec_stat=december_df.describe()
percipitation_june_stat=june_prcp_df.describe()
percipitation_dec_stat=december_prcp_df.describe()

temp_summary_df=temps_june_stat.merge(temps_dec_stat,left_index=True, right_index=True).\
    merge(percipitation_june_stat, left_index=True, right_index=True).\
    merge(percipitation_dec_stat, left_index=True, right_index=True)
temp_summary_df
```
